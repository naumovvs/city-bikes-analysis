```python
import csv
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import Birch
from sklearn.cluster import AgglomerativeClustering
```

```python
trips = []

with open('data/trips_clean_all.txt', 'r') as csvfile:
    trips_reader = csv.reader(csvfile, delimiter='\t', quotechar='.')
    for data in trips_reader:
        # 1 - trip duration, 2 - travel distance, 
        # 3 - idle time, 4 - elongation coefficient
        trips.append([float(data[1]), float(data[2]), float(data[3]), float(data[4])])
```

```python
def cluster_analysis_KMeans(trips, ys_idx, k=2):
    kmeans = KMeans(n_clusters=k).fit(trips)
    #kmeans = MiniBatchKMeans(n_clusters=k).fit(trips)
    #kmeans = Birch(n_clusters=k).fit(trips) # ! k=2
    #kmeans = AgglomerativeClustering(n_clusters=k).fit(trips) 
    clusters = [[] for _ in range(k)]
    for i in range(len(trips)):
        clusters[kmeans.labels_[i]].append(i)
    counts = [0 for _ in range(k)]
    for y in ys_idx:
        for i in range(k):
            if y in clusters[i]:
                counts[i] += 1
                break
    return counts, [len(c) for c in clusters]

def cluster_analysis_MiniBatchKMeans(trips, ys_idx, k=2):
    kmeans = MiniBatchKMeans(n_clusters=k).fit(trips)
    #kmeans = Birch(n_clusters=k).fit(trips) # ! k=2
    #kmeans = AgglomerativeClustering(n_clusters=k).fit(trips) 
    clusters = [[] for _ in range(k)]
    for i in range(len(trips)):
        clusters[kmeans.labels_[i]].append(i)
    counts = [0 for _ in range(k)]
    for y in ys_idx:
        for i in range(k):
            if y in clusters[i]:
                counts[i] += 1
                break
    return counts, [len(c) for c in clusters]

def cluster_analysis_Birch(trips, ys_idx, k=2):
    kmeans = Birch(n_clusters=k).fit(trips) # ! k=2
    #kmeans = AgglomerativeClustering(n_clusters=k).fit(trips) 
    clusters = [[] for _ in range(k)]
    for i in range(len(trips)):
        clusters[kmeans.labels_[i]].append(i)
    counts = [0 for _ in range(k)]
    for y in ys_idx:
        for i in range(k):
            if y in clusters[i]:
                counts[i] += 1
                break
    return counts, [len(c) for c in clusters]

def cluster_analysis_AgglomerativeClustering(trips, ys_idx, k=2):
    kmeans = AgglomerativeClustering(n_clusters=k).fit(trips) 
    clusters = [[] for _ in range(k)]
    for i in range(len(trips)):
        clusters[kmeans.labels_[i]].append(i)
    counts = [0 for _ in range(k)]
    for y in ys_idx:
        for i in range(k):
            if y in clusters[i]:
                counts[i] += 1
                break
    return counts, [len(c) for c in clusters]
```

```python
ks = range(2, 21)
n, m = float(len(trips)), float(len(trip_attr_indices))
deltas1, deltas2, deltas3, deltas4 = [], [], [], []
for k in ks:
    cc1 = cluster_analysis_KMeans(trips, trip_attr_indices, k)
    cc2 = cluster_analysis_MiniBatchKMeans(trips, trip_attr_indices, k)
    cc3 = cluster_analysis_Birch(trips, trip_attr_indices, k)
    cc4 = cluster_analysis_AgglomerativeClustering(trips, trip_attr_indices, k)
    res1 = (np.array(cc1[0]) / m) - (np.array(cc1[1]) / n)
    res2 = (np.array(cc2[0]) / m) - (np.array(cc2[1]) / n)
    res3 = (np.array(cc3[0]) / m) - (np.array(cc3[1]) / n)
    res4 = (np.array(cc4[0]) / m) - (np.array(cc4[1]) / n)
    deltas1.append(res1.max())
    deltas2.append(res2.max())
    deltas3.append(res3.max())
    deltas4.append(res4.max())    
    print str(res1.argmax()) + " / " + str(k) + ": " + str(res1.max())
    print str(res2.argmax()) + " / " + str(k) + ": " + str(res2.max())
    print str(res3.argmax()) + " / " + str(k) + ": " + str(res3.max())
    print str(res4.argmax()) + " / " + str(k) + ": " + str(res4.max())
```

```python
plt.figure(dpi=150)
plt.xlim(2, 20)
plt.xlabel('Number of clusters')
plt.ylabel('Objective function value')
plt.plot(ks, deltas1, label='KMeans')
plt.plot(ks, deltas2, label='MiniBatchKMeans')
plt.plot(ks, deltas3, label='Birch')
plt.plot(ks, deltas4, label='AgglomerativeClustering')
plt.legend(loc='upper right')
plt.show()
```
